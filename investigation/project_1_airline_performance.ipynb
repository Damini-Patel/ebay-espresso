```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
import seaborn as sns

# Flight data files
data_path = "../resources/cleansheet1_with_airline.csv"

# Read the flight data
flight_airline_data = pd.read_csv(data_path)
flight_airline_data

```

```python
#Remove commas in figures and convert to float
clean_flight_airline_data = flight_airline_data
clean_flight_airline_data["Passengers In"]= clean_flight_airline_data["Passengers In"].str.replace(",","").replace("..","0").astype(float)

clean_flight_airline_data

```

```python
#Remove commas in figures and convert to float
clean_flight_airline_data["Passengers Out"]= clean_flight_airline_data["Passengers Out"].str.replace(",","").replace("..","0").astype(float)
clean_flight_airline_data
```

```python
#Replace fields containing no data '..' with 0
clean_flight_airline_data["Mail In"]= clean_flight_airline_data["Mail In"].str.replace("..","0").astype(float)
clean_flight_airline_data["Mail Out"]= clean_flight_airline_data["Mail Out"].str.replace("..","0").astype(float)
clean_flight_airline_data["Freight In"]= clean_flight_airline_data["Freight In"].str.replace(",","").replace("..","0").astype(float)
clean_flight_airline_data["Freight Out"]= clean_flight_airline_data["Freight Out"].str.replace(",","").replace("..","0").astype(float)
clean_flight_airline_data
```

```python
#Check column types to verify data types and conversion
clean_flight_airline_data.info()
```

```python
#Calculate Passenger, Mail & Freight In and Out numbers
clean_flight_airline_data["Passengers_Total"] = clean_flight_airline_data["Passengers In"] + clean_flight_airline_data["Passengers Out"]
clean_flight_airline_data["Freight_Total"] = clean_flight_airline_data["Freight In"] + clean_flight_airline_data["Freight Out"]
clean_flight_airline_data["Mail_Total"] = clean_flight_airline_data["Mail In"] + clean_flight_airline_data["Mail Out"]

#Combine Freight and Mail totals for easy comparison against passenger numbers
clean_flight_airline_data["Freight_Mail_Total"] = clean_flight_airline_data["Mail_Total"] + clean_flight_airline_data["Freight_Total"]
clean_flight_airline_data

```

```python
#Remove unused columns
clean_flight_airline_data = clean_flight_airline_data.drop(columns=["Country to/from","Freight In","Freight Out", "Mail In", "Mail Out", "Passengers In", "Passengers Out", "Freight_Total", "Mail_Total"])
clean_flight_airline_data
```

```python

# grouped_airline_flight_df = clean_flight_airline_data.groupby(["Scheduled Operator","Month","Year" ], as_index=False)
grouped_airline_flight_df = clean_flight_airline_data.drop(columns="Year")
grouped_airline_flight_df = grouped_airline_flight_df.groupby(["Scheduled Operator"])

#Get top airlines by total Freight and Mail weight
# top_freight_mail_airlines_df = grouped_airline_flight_df.sum().sort_values("Freight_Mail_Total", ascending=False)[:5]
top_freight_mail_airlines_df = grouped_airline_flight_df.sum().sort_values("Freight_Mail_Total", ascending=False)
top_freight_mail_airlines_df["Freight_Mail_Total"]

```

```python
### Observation: Singaporean Airlines carried the most freight ove the last 10yrs closely followed by Qantas Airlines
top_freight_mail_airlines_df["Freight_Mail_Total"].plot.bar(alpha=0.7, colormap="summer")
plt.legend(loc="best")
plt.title("Top Airlines by Freight & Mail")
plt.xlabel("Top Airlines")
plt.ylabel("Weight (Tonnes)")
```

```python
# Save list of top airlines by Freight and Mail weight (tonnes) since 2009
top_freight_mail_airlines_list = top_freight_mail_airlines_df.index
top_freight_mail_airlines_list
```

```python
#Get top airlines by total number of passengers travelled since 2009
# top_passenger_airlines_df = grouped_airline_flight_df.sum().sort_values("Passengers_Total", ascending=False)[:5]
top_passenger_airlines_df = grouped_airline_flight_df.sum().sort_values("Passengers_Total", ascending=False)
top_passenger_airlines_df["Passengers_Total"]
top_passenger_airlines_df["Passengers_Total"].plot.bar(alpha=0.7, colormap="winter")
plt.legend(loc="best")
plt.title("Top Airlines by Passengers")
plt.xlabel("Top Airlines")
plt.ylabel("Passengers (Millions)")

```

```python
### Observation: Quantas Airlines carries at least double the number passengers compared to any other airline.
# Save list of top airlines by passengers carried
top_passenger_airlines_list = top_passenger_airlines_df.index
top_passenger_airlines_list
```

```python
#Combine top airlines for freight-mail and passengers
top_airlines_combined_list = top_passenger_airlines_list.append(top_freight_mail_airlines_list)

top_airlines_combined_list = top_airlines_combined_list.unique()
top_airlines_combined_list
```

```python
# Filter top airline data using list of top airlines i.e. highest cargo and passenger to date
top_airlines_flight_data_df = clean_flight_airline_data[clean_flight_airline_data["Scheduled Operator"].isin(top_airlines_combined_list)]
#Filter by data later than 2019
top_airlines_flight_data_df = top_airlines_flight_data_df.loc[top_airlines_flight_data_df.Year >= 2019]
# top_airlines_flight_data_df = top_airlines_flight_data_df.drop(columns="Year")
top_airlines_flight_data_df
```

```python
# Filter top airline data using list of top airlines i.e. highest cargo and passenger to date
top_airlines_flight_data_df = clean_flight_airline_data[clean_flight_airline_data["Scheduled Operator"].isin(top_airlines_combined_list)]
#Filter by data later than 2019
top_airlines_flight_data_df = top_airlines_flight_data_df.loc[top_airlines_flight_data_df.Year >= 2019]
# top_airlines_flight_data_df = top_airlines_flight_data_df.drop(columns="Year")
top_airlines_flight_data_df
```

```python
#Dataframe for comparison between 2020 vs. 2019
change_df = top_airlines_flight_data_df.groupby(["Scheduled Operator", "Year"], as_index = True)
change_df2 = pd.DataFrame(change_df.sum())
change_df2
```

```python
#Calculate change and save into list
passengers_totals_list = change_df2["Passengers_Total"].tolist()

#Calculate total passengers (in and out) % change between 2019 vs 2020
passenger_change_list = [round((passengers_totals_list[i+1]-passengers_totals_list[i]) / passengers_totals_list[i],2)*100  1 for i in range(len(passengers_totals_list)-1)]

# Discard perecentage result from different airline
passenger_percentage_change_list = passenger_change_list[::2] #ignores every second element in the list
passenger_percentage_change_list
```

```python
#Calculate change and save into list
freight_mail_totals_list = change_df2["Freight_Mail_Total"].tolist()

#Calculate total passengers (in and out) % change between 2019 vs 2020
freight_mail_list = [round((freight_mail_totals_list[i+1]-freight_mail_totals_list[i]) / freight_mail_totals_list[i],2)*100 for i in range(len(freight_mail_totals_list)-1)]

# Discard perecentage result from different airline
freight_mail_percentage_change_list = freight_mail_list[::2] #ignores every second element in the list
freight_mail_percentage_change_list
```

```python
# Create new dataframe comparing passenger and cargo numbers between 2019 vs 2020
top_airlines_combined_list = top_airlines_combined_list.sort_values() #Arrange alphabetically to ensure results match up
airline_strategy_change_df = pd.DataFrame({
    "Scheduled_Operator":top_airlines_combined_list,
    "Change_in_Passengers":passenger_percentage_change_list,
    "Change_in_Freight_Mail":freight_mail_percentage_change_list
    })

airline_strategy_change_df
```

```python
#Plot Change in Passengers
fig, ax = plt.subplots(figsize=(30,10)) 
plt.bar(top_airlines_combined_list, passenger_percentage_change_list,  color="b", align="center")
plt.title("2020 vs 2019 Change in Passengers",  fontsize=30)
plt.xlabel("Top Airlines", fontsize=25)
plt.ylabel("Percentage (%)", fontsize=25)
plt.xticks(ticks=[0,1,2,3,4,5],labels=top_airlines_combined_list, fontsize=25)
plt.yticks(fontsize=20)
```

```python
### Observation: All airlines dropped passenger travel by an average of 70%. Cathay Pacific had the largest change in number of passengers flying at -74%. But marginally above average
```

```python
#Plot Change in Cargo
fig, ax = plt.subplots(figsize=(30,10))
plt.bar(top_airlines_combined_list, freight_mail_percentage_change_list, color="g", align="center")
plt.title("2020 vs 2019 Change in Freight & Mail", fontsize=30)
plt.xlabel("Top Airlines", fontsize=25)
plt.ylabel("Percentage (%)", fontsize=25)
plt.xticks(ticks=[0,1,2,3,4,5],labels=top_airlines_combined_list, fontsize=25)
plt.yticks(fontsize=20)
```

```python
#Alternative Bar Cluster Plot Visualisation
matplotlib.style.use('ggplot')
df = airline_strategy_change_df
fig, ax = plt.subplots(figsize=(30,10))
df[['Change_in_Passengers']].plot.bar(stacked=True, width=0.2, position=1.5, colormap="winter", ax=ax, alpha=0.7)
df[['Change_in_Freight_Mail']].plot.bar(stacked=True, width=0.2, position=0.5, colormap="summer", ax=ax, alpha=0.7)
plt.legend(loc="best", fontsize=18)
plt.title("2020 vs 2019 Percentage Decrease in Performance", fontsize=30)
plt.xlabel("Top Airlines", fontsize=25)
plt.ylabel("Percentage (%)", fontsize=25)
plt.xticks(ticks=[0,1,2,3,4,5],labels=top_airlines_combined_list, fontsize=25)
plt.yticks(fontsize=20)
plt.show()
```
